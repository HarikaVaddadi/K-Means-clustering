**K-Means clustering on IRIS Dataset**

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
```

```
iris = datasets.load_iris()
```

```
print(iris.data)
```

```
print(iris.target_names)
```

```
print(iris.target)
```

Import data to the Pandas Data Frame

```
x = pd.DataFrame(iris.data, columns=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
y = pd.DataFrame(iris.target, columns=['Target'])
```

```
x.head()
```

```
y.head()
```

**Let us visualize the data through a scatter plot.**

```
colors = np.array(['red', 'green', 'blue'])
plt.scatter(x['Sepal Length'], x['Sepal Width'],c=colors[y['Target']])
plt.xlabel('Sepa1 Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width',fontsize=16)
```

```
plt.scatter(x['Petal Length'], x['Petal Width'],c=colors[y['Target']])
plt.xlabel('Peta1 Length')
plt.ylabel('Petal Width')
plt.title('Petal Length vs Petal Width',fontsize=16)
```

**Model training and evaluation**

```
km = KMeans(n_clusters = 3)
km.fit(x)
new_label=km.labels_
```

```
centers = km.cluster_centers_
print(centers)
```

**Lets compare our original data versus predicted clustered results**

```
plt.figure(figsize=(12,4))
plt.subplot(1, 2, 1)
plt.scatter(x['Sepal Length'], x['Sepal Width'], c=colors[y['Target']])
plt.xlabel('SePa1 Length')
plt.ylabel('Sepal Width')
plt.title('Actual clusters',fontsize=16)

plt.subplot(1, 2, 2)
plt.scatter(x['Sepal Length'], x['Sepal Width'], c=colors[new_label])
plt.xlabel('SePa1 Length')
plt.ylabel('Sepal Width')
plt.title("Predicted clusters",fontsize=16)
```

```
plt.figure(figsize=(12,4))
plt.subplot(1, 2, 1)
plt.scatter(x['Petal Length'], x['Petal Width'], c=colors[y['Target']])
plt.xlabel('Peta1 Length')
plt.ylabel('Petal Width')
plt.title('Actual clusters',fontsize=16)

plt.subplot(1, 2, 2)
plt.scatter(x['Petal Length'], x['Petal Width'], c=colors[new_label])
plt.xlabel('Peta1 Length')
plt.ylabel('Petal Width')
plt.title('Predicted clusters',fontsize=16)
```

```
#Accurary of the model
import sklearn.metrics as sm
sm.accuracy_score(new_label, y['Target'])
```

Here accuracy_score comes out to be 0.89, it means our model works with 89% accuracy.(To be more clear, 89% means 89 correct predictions out of 100 total cases.)
